```python
# Obtener HTML de una dirección web.
from urllib.request import Request, urlopen
#url = 'https://www.culturabai.com/collections/all'
url = 'https://tiendadeaventuracr.com/collections/sombreros-para-senderismo-o-playa-sombreros-para-el-sol'
req = Request(url , headers={'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(req).read()
pagina = BeautifulSoup(webpage, 'html.parser')
```

```python
#d) Determinar si el producto esta agotado.
# Ningún producto está agotado, se asignó una columna con un NO para todos los productos (d)
```

```python
#e) Muestre al final la tabla de datos. Se muestra acontinuación:
```

```python

productos = pagina.find_all("div", {'class':'productitem'})

tablaProductos = pd.DataFrame()

for producto in productos:
  htmlNombre = producto.find_all("h2", {'class':'productitem--title'})   #(a)
  if len(htmlNombre) > 0:
    etqNombre = htmlNombre[0].find("a").text
    
  htmlPrecio = producto.find_all("div", {'class':'price productitem__price'}) #(b)
  if len(htmlPrecio) > 0:
    etqPrecioOriginal = htmlPrecio[0].text
  else:
    etqPrecioOriginal = None
    
  htmlPrecioDescuento = producto.find_all("div", {'class':'price__current price__current--emphasize price__current--on-sale'}) #(c)
  if len(htmlPrecioDescuento) > 0:        
    etqPrecioDescuento = htmlPrecioDescuento[0].text
  else:
    etqPrecioDescuento = None  
    
  nuevoProducto = pd.DataFrame({"Nombre" : [etqNombre], "precioOriginal" : etqPrecioOriginal,
                                  "PrecioCondescuento" : [etqPrecioDescuento], "Agotado": "NO"})  #(d)  
  tablaProductos = pd.concat([tablaProductos, nuevoProducto])
print(tablaProductos)
```
